```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error

train_size = 50
test_size = 20
```

隨機產生train_X test_x
Sin函數產生train_y test_y

```python
train_X = np.random.uniform(low=-1, high=2, size=train_size)
test_X = np.random.uniform(low=-2, high=3, size=test_size)
train_y = np.sin(train_X)+np.random.normal(loc=0, scale=0.1, size=train_size)
test_y = np.sin(test_X)+np.random.normal(loc=0, scale=0.1, size=test_size)
```

(資料分布)

```python
plt.scatter(train_X, train_y, label='Train')  # Train data in blue circles
plt.scatter(test_X, test_y, label='Test', color='red')  # Test data in red squares

plt.xlabel('X')
plt.ylabel('y')
plt.title('Train and Test Data')
plt.legend()  # Show legend to distinguish train and test data
plt.show()
```

以6次方程式當model
fit_transform()
predict()

結果：MSE

```python
poly = PolynomialFeatures(degree=3)
train_poly_X = poly.fit_transform(train_X.reshape(train_size, 1))
test_poly_X = poly.fit_transform(test_X.reshape(test_size, 1))

model = Ridge(alpha=1.)
model.fit(train_poly_X, train_y)
train_y_pred = model.predict(train_poly_X)
test_y_pred = model.predict(test_poly_X)

print(mean_squared_error(train_y, train_y_pred))
print(mean_squared_error(test_y, test_y_pred))
```

結果以圖呈現（train、test）

```python
plt.scatter(test_X, test_y, label='Train')  # Train data in blue circles
plt.scatter(test_X, test_y_pred, label='Test', color='red')  # Test data in red squares


plt.xlabel('X')
plt.ylabel('y')
plt.title('Train and Test Data')
plt.legend()  # Show legend to distinguish train and test data
plt.show()
```
